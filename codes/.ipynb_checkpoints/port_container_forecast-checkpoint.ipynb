# Port of Vancouver Containers Statistics Forecast

## Summary

- Objective: forecast future month TEU of Port of Vancouver to give guidence in the future.
- Data Source: [container statistics monthly 2008 - 2024](https://www.portvancouver.com/media/documents/container-statistics-monthly-2008-2024)
- Approach: Meta Prophet
- Setup: 
    - Variable (y) : TEU Count total empty
    - Frequency: monthly
    - Training data: before 2024
    - Testing data: after and include 2024-01-01.
    - Evaluation metrics: mean absolute error (\text{MAE})
- Procedures and results
    1. Model 0 default prophet model with all training data, $\text{MAE}_0 = 9260$. In section [Observation from Model 0], determine possible reasons of deviations, such as pandemic and multiplicative seasonality. Add the modelling plan for the following few models.
    ![model0](../figures/Model_0.png)
    2. Model 1 default prophet model without pandemic data, $\text{MAE}_1 = 8400$. Deviation is reduced by ignoring pandemic data, but the model is still underfitting. Post-pandemic effect exists.
    ![model1](../figures/Model_1.png)
    3. Model 2 with pandemic data and fine-tuning multiple parameters, $\text{MAE}_2 = 7421$. Evaluate the cross-validation result in training dataset by parameter frequency, time weighted cross-validation MAE average and MAE at testing dataset. Model overfitts during pandemic, which requires further modelling of pandemic data.
    ![model2](../figures/Model_2.png)
    4. Model 3 with pandemic data, fine-tuned parameter `holidays_prior_scale`, and extra regressors, BC COVID-19 stringency index, $\text{MAE}_3 = 6744$. Model 3 overcomes the problem of overfitting and learns the peak from March to August very well with modified yearly seasonality. But there are also two major deviations, January to Februray and September to November, which requires more investigations.
    ![model3](../figures/Model_3.png)
    ![model3_components](../figures/Model_3_components.png)

### Observation from Model 0

The default prophet model without any tuning has two major deviations, under estimation between 2020 - 2022 and over estimation during 2023. The period of the abnormal trend matches the duration of COVID-19. There are more labour disruptions, congestion at port and altered consumer behavior according to [Transport Canada's 2021 report](https://tc.canada.ca/en/corporate-services/transparency/briefing-documents-transport-canada/2021/current-topics/canada-s-freight-transportation-system-global-crisis-container-shipping-supply-chains). It is crucial to separate the pandemic data in the following prediction.

According to global market and local policy, the pandemic data are influenced mainly during the following two overlapping periods:

1. [COVID19 Pandemic period in BC](https://en.wikipedia.org/wiki/COVID-19_pandemic_in_British_Columbia), from March 18, 2020 to June 30, 2021. Lockdown in BC resulted in series container congestions at the Port of Vancouver, high container rates and  high demand for containerized goods.
2. [Top trading economy](chrome-extension://efaidnbmnnnibpcajpcglclefindmkaj/https://www.portvancouver.com/sites/default/files/2024-08/2022-Stats-Overview_0.pdf), [COVID19 pandemic period in China](https://en.wikipedia.org/wiki/COVID-19_pandemic_in_mainland_China#December_2022%E2%80%93January_2023_surge), from January 23, 2020 to January 8, 2023. Containered cargo unit counts are influenced on both inbound and outbound goods.

Unlike the Prophet offical guide of [Handling Shocks](https://facebook.github.io/prophet/docs/handling_shocks.html), there is a lack of granularity and dimensions, e.g., the trading economy of exporting and importing. So, we will use the monthly data for modeling.

1. BC_pandemic: 2020 March - 2021 June
2. China_pandemic: 2020 January - 2022 December 

According to the trend, it is possible that this trend has a multiplicative seasonality. Also, there are potential changepoints not included during pandemic periods.

Three models will be built:
1. Training the model without pandemic data. Directly predict the future.
2. Training the model with pandemic data with fine tuned parameters.
3. Training the model with pandemic data with extra modeling during pandemic, e.g. .

For the second approach, cross-validation will be applied to tuning the optimal scales for all parameters, such as holiday scale, change point range, seasonality scale and seasonality mode. All models will be compared in testing data using cross-validation with the following sliding-window settings according to [Prophet guide of cross-validation](https://facebook.github.io/prophet/docs/diagnostics.html#cross-validation) and [cross-validation for time series](https://robjhyndman.com/hyndsight/tscv/):

- assess prediction performance on a horizon of 12 months
- start with 6 years of training data in the first cutoff 
- make predictions every 6 months

### Model 2 - Fine Tuning Parameters + all data

#### Split Training and Testing

```python
predict_column = 'TEU COUNT_Total_Empty' # predict one variable first.

test_year = 2024
training_index = [i for i in month_series.index if month_series['year'][i] != str(test_year)]
testing_index = [i for i in month_series.index if month_series['year'][i] == str(test_year)]

data = copy.deepcopy(month_series[['ds', predict_column]])
data.rename(columns={predict_column: 'y'}, inplace=True)

training_data = copy.deepcopy(data.loc[training_index])
testing_data = copy.deepcopy(data.loc[testing_index])
```

#### Modelling

```python
param_grid = {  
    'changepoint_prior_scale': [0.01, 0.1, 0.5],
    'seasonality_prior_scale': [0.01, 0.1, 1.0, 10.0],
    'holidays_prior_scale': [0.01, 1.0, 5, 10.0],
    'seasonality_mode': ['additive', 'multiplicative']
}

# keep change point including pandemic period
# change point range is set to be before 2023.
changepoint_range = training_data[training_data['ds'] == '2023-01-01'].index[0] / training_data.shape[0]

pandemic_df = pd.DataFrame([
    {'holiday': 'BC_pandemic', 'ds': '2020 March', 'lower_window': 0, 'ds_upper': '2021 June'},
    {'holiday': 'China_pandemic', 'ds': '2020 January', 'lower_window': 0, 'ds_upper': '2022 December'},
])
for t_col in ['ds', 'ds_upper']:
    pandemic_df[t_col] = pd.to_datetime(pandemic_df[t_col], format='%Y %B')
pandemic_df['upper_window'] = (pandemic_df['ds_upper'] - pandemic_df['ds']).dt.days
pandemic_df
```

#### Training by Cross-Validation

```python
# Generate all combinations of parameters
all_params = [dict(zip(param_grid.keys(), v)) for v in itertools.product(*param_grid.values())]
rmses = {x+1:[] for x in range(12)}  # Store the RMSEs for each params here
mae = {x+1:[] for x in range(12)} # MAE

# Use cross validation to evaluate all parameters
for params in all_params:
    print(params)
    m = Prophet(changepoint_range = changepoint_range, holidays=pandemic_df, 
                yearly_seasonality=True, weekly_seasonality=False, daily_seasonality=False, 
                **params).fit(training_data)  # Fit model with given params
    tmp_df_cv = cross_validation(m, initial=str(365 * 8) + ' days', period=str(30 * 6) + ' days', horizon = '365 days', disable_tqdm=False, parallel=None)
    tmp_df_p = performance_metrics(tmp_df_cv, monthly=True, rolling_window=0.08)
    for i in range(tmp_df_p.shape[0]):
        rmses[i+1].append(tmp_df_p['rmse'].values[i])
        mae[i+1].append(tmp_df_p['mae'].values[i])
```

#### Cross-Validation Result

```python
# collect all results
rmses_df = pd.DataFrame(rmses)
rmses_df.rename({i+1: 'rmse_month_' + str(i+1) for i in range(12)}, axis=1, inplace=True)

mae_df = pd.DataFrame(mae)
mae_df.rename({i+1: 'mae_month_' + str(i+1) for i in range(12)}, axis=1, inplace=True)

tuning_results = pd.DataFrame(all_params)
tuning_results = pd.concat([tuning_results, rmses_df, mae_df], axis=1)

# list the best parameters
best_tuning_para_idx = tuning_results[rmses_df.columns.to_list() + mae_df.columns.to_list()].idxmin(axis=0)
best_metric = []
best_params = []

for i in range(len(best_tuning_para_idx)):
    metric = best_tuning_para_idx.index[i]
    idx = best_tuning_para_idx[i]
    metric_name = metric.split('_')[0]
    month = int(metric.split('_')[-1])
    best_params.append(tuning_results.loc[idx][:4].to_dict())
    best_metric.append({'metric': metric_name,
                        'month': month,
                        'value': tuning_results.loc[idx, metric]
                       })
best_param_result = pd.concat([pd.DataFrame(best_params), pd.DataFrame(best_metric)], axis=1)
best_param_result['best_para_idx'] = best_tuning_para_idx.to_list()


with pd.ExcelWriter(predict_column + '_' + 'tuning_results.xlsx') as excel_writer:
    # use to_excel function and specify the sheet_name and index 
    # to store the dataframe in specified sheet
    tuning_results.to_excel(excel_writer, sheet_name = 'results', float_format="%.4f", header=True, index=False)
    best_param_result.to_excel(excel_writer, sheet_name = 'best_results', float_format="%.4f", header=True, index=False)

best_param_result
```

Both MAE and RMSE have same optimal model choice for one to four month forecasting. There are different parameter combinations in different month prediction result, with the most frequent one (combination No. 43), `seasonality_mode='multiplicative'`, `changepoint_prior_scale=0.1`, `seasonality_prior_scale=0.1`, `holidays_prior_scale=1`. One approach is to establish multiple models for different prediction horizon. For simplicity, to find the best parameters, another approach is to compare cross-validation result with weighted averaged metric. Each metric is denoted $\omega_i$ with $i$ as future months. Then we have the following weighted average $\Omega$ with inversely proportional weights $1-\frac{i-1}{12}$ to match the time decaying:
$$\Omega = \sum_{i=1}^{12}(1-\frac{i-1}{12})\omega_i.$$

With an additional weight, it will be more flexibile to choose best parameters with various business plan. For example, we need a quarterly overall best forecast, so we can choose a quaterly weights, $1 - \frac{\lceil i/3 \rceil-1}{4}$.

#### Choose Best Model with Weighted Average

```python
def weighted_average_metric(x, metric, periods=12, freq='month', weight='proportional'):
    metric_col = [metric + '_' + freq + '_' + str(period+1) for period in range(periods)]
    if weight=='proportional':
        weights = [1 - i/periods for i in range(periods)]

    weighted_average = sum([x[metric_col[i]] * weights[i] for i in range(periods)])
    
    return(weighted_average)

weighted_mae = tuning_results.loc[best_param_result['best_para_idx'].unique()].apply(lambda x: weighted_average_metric(x, metric='mae'), axis=1)
weighted_rmse = tuning_results.loc[best_param_result['best_para_idx'].unique()].apply(lambda x: weighted_average_metric(x, metric='rmse'), axis=1)
only_one_best_df = pd.DataFrame({
    'w_avg_mae': weighted_mae,
    'w_avg_rmse' : weighted_rmse
})
# only_one_best_df.reset_index(inplace=True)
params_name = list(param_grid.keys())
only_one_best_df = pd.concat([only_one_best_df, tuning_results.loc[only_one_best_df.index, params_name]], axis=1, ignore_index=False)
```

### Model 3 - Extra Regressor

#### Import Stringency Index Data

```python
path_name = './data/port_of_vancouver'

stringency_file_name = 'BC_COVID-19_STRINGENCY_INDEX.csv'
#reads table from pdf file
stringency_df = pd.read_csv(path_name + '/' + stringency_file_name, header=0)

stringency_df['ds'] = pd.to_datetime(stringency_df['date'])
stringency_df['year_month'] = stringency_df['ds'].dt.strftime('%Y-%m-01')
stringency_df_agg = stringency_df.groupby(['year_month']).agg({'index': np.average})
stringency_df_agg.reset_index(inplace=True, names='ds')
stringency_df_agg['ds']  = pd.to_datetime(stringency_df_agg['ds'])

stringency_df_agg.plot(x='ds', y='index', title='BC COVID-19 Stringency Index - Aggregated in Month')
```

#### Modify Training and Testing Data

```python
training_data_index = training_data.join(stringency_df_agg.set_index('ds'), on='ds', how='left')
training_data_index.fillna(0, inplace=True)
testing_data_index = testing_data.join(stringency_df_agg.set_index('ds'), on='ds', how='left')
testing_data_index.fillna(0, inplace=True)
```

#### Modelling

```python
param_grid_2 = {
    'holidays_prior_scale': [0.01, 0.05, 0.1, 0.5, 1.0, 5.0, 10.0]
}

# keep change point including pandemic period
# change point range is set to be before 2023.
changepoint_range = training_data[training_data['ds'] == '2023-01-01'].index[0] / training_data.shape[0]

pandemic_df = pd.DataFrame([
    {'holiday': 'BC_pandemic', 'ds': '2020 March', 'lower_window': 0, 'ds_upper': '2021 June'},
    {'holiday': 'China_pandemic', 'ds': '2020 January', 'lower_window': 0, 'ds_upper': '2022 December'},
])
for t_col in ['ds', 'ds_upper']:
    pandemic_df[t_col] = pd.to_datetime(pandemic_df[t_col], format='%Y %B')
pandemic_df['upper_window'] = (pandemic_df['ds_upper'] - pandemic_df['ds']).dt.days
pandemic_df
```

```python
# Generate all combinations of parameters
all_params = [dict(zip(param_grid_2.keys(), v)) for v in itertools.product(*param_grid_2.values())]
rmses_2 = {x+1:[] for x in range(12)}  # Store the RMSEs for each params here
mae_2 = {x+1:[] for x in range(12)} # MAE

# Use cross validation to evaluate all parameters
for params in all_params:
    print(params)
    m = Prophet(changepoint_range = changepoint_range, holidays=pandemic_df, 
                yearly_seasonality=True, weekly_seasonality=False, daily_seasonality=False, 
                seasonality_mode='multiplicative', changepoint_prior_scale=0.1, seasonality_prior_scale=0.1,
                **params)
    m.add_regressor('index')
    m.fit(training_data_index)  # Fit model with given params
    tmp_df_cv = cross_validation(m, initial=str(365 * 8) + ' days', period=str(30 * 6) + ' days', horizon = '365 days', disable_tqdm=False, parallel=None)
    tmp_df_p = performance_metrics(tmp_df_cv, monthly=True, rolling_window=0.08)
    for i in range(tmp_df_p.shape[0]):
        rmses_2[i+1].append(tmp_df_p['rmse'].values[i])
        mae_2[i+1].append(tmp_df_p['mae'].values[i])

```

#### Find Best Parameters

```python
# collect all results
rmses_df_2 = pd.DataFrame(rmses_2)
rmses_df_2.rename({i+1: 'rmse_month_' + str(i+1) for i in range(12)}, axis=1, inplace=True)

mae_df_2 = pd.DataFrame(mae_2)
mae_df_2.rename({i+1: 'mae_month_' + str(i+1) for i in range(12)}, axis=1, inplace=True)

tuning_results_2 = pd.DataFrame(all_params)
tuning_results_2 = pd.concat([tuning_results_2, rmses_df_2, mae_df_2], axis=1)

# list the best parameters
best_tuning_para_idx_2 = tuning_results_2[rmses_df_2.columns.to_list() + mae_df_2.columns.to_list()].idxmin(axis=0)
best_metric_2 = []
best_params_2 = []

for i in range(len(best_tuning_para_idx_2)):
    metric = best_tuning_para_idx_2.index[i]
    idx = best_tuning_para_idx_2[i]
    metric_name = metric.split('_')[0]
    month = int(metric.split('_')[-1])
    best_params_2.append(tuning_results_2.loc[idx][:4].to_dict())
    best_metric_2.append({'metric': metric_name,
                        'month': month,
                        'value': tuning_results_2.loc[idx, metric]
                       })
best_param_result_2 = pd.concat([pd.DataFrame(best_params_2), pd.DataFrame(best_metric_2)], axis=1)
best_param_result_2['best_para_idx'] = best_tuning_para_idx_2.to_list()

best_param_result_2
```

The MAE in testing dataset decreases from 7421 to 6744, which is the minimum among all models.
